```python
# import xmf-en.tsv
import csv


pairs = []

# read the file
with open('xmf-en.tsv', 'r') as file:
    # print only rows with two items (tab separated )
    for line in file:
        # create a list of items, remove the newline character
        items = line.split('\t')
        items = [item.strip() for item in items]
        if items and len(items) == 2 and items[1] != '':
            print(items)
            pairs.append(items)

print(len(pairs))
```
